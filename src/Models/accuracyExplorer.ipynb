```python
import pandas as pd
import json
```

```python
# Carregar o arquivo JSON
with open('testModelLog58.json', 'r') as file:
    data = json.load(file)

# Filtrar os dados com base nos parâmetros desejados
filtered_data = [item for item in data if item.get('params') == patternFrame]

# Criar um DataFrame pandas com os dados filtrados
df = pd.DataFrame(filtered_data)
print(df)
```

```python
# Criar um novo DataFrame com as colunas desejadas
result_df = pd.DataFrame({
    'Índice': df.index,
    'Data': df['executionDate'],
    'Acurácia de Treino': [accuracy[0] for accuracy in df['accuracy']],
    'Acurácia de Teste': [accuracy[1] for accuracy in df['accuracy']]
})

# Salvar o DataFrame em um arquivo CSV
result_df.to_csv('resultado.csv', index=False)
```
